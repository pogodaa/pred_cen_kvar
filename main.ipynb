```python
import cianparser
import pandas as pd
from time import sleep

def get():
      moscow_parser = cianparser.CianParser(location='Москва')
      moscow_oblast_parser = cianparser.CianParser(location='Московская область')
      rostov_parser = cianparser.CianParser(location='Ростов-на-Дону') 
      rostov_oblast_parser = cianparser.CianParser(location='Ростовская область')
      krasnodar_parser = cianparser.CianParser(location='Краснодар')
      krasnodar_oblast_parser = cianparser.CianParser(location='Краснодарский край')

      sleep(10)
      data = moscow_parser.get_flats(deal_type="sale", rooms=(1), with_saving_csv=True, additional_settings={"start_page":1, "end_page":2})
      sleep(20)
      data += moscow_oblast_parser.get_flats(deal_type="sale", rooms=(1), with_saving_csv=True, additional_settings={"start_page":1, "end_page":2})
      sleep(20)
      data += rostov_parser.get_flats(deal_type="sale", rooms=(1), with_saving_csv=True, additional_settings={"start_page":1, "end_page":2})
      sleep(20)
      data += rostov_oblast_parser.get_flats(deal_type="sale", rooms=(1), with_saving_csv=True, additional_settings={"start_page":1, "end_page":2})
      sleep(20)
      data += krasnodar_parser.get_flats(deal_type="sale", rooms=(1), with_saving_csv=True, additional_settings={"start_page":1, "end_page":2})
      sleep(20)
      data += krasnodar_oblast_parser.get_flats(deal_type="sale", rooms=(1), with_saving_csv=True, additional_settings={"start_page":1, "end_page":2})
      sleep(10)

      df_moscow = pd.read_csv('cian_flat_sale_1_2_moskva_28_Oct_2024_10_*.csv')
      df_moscow_oblast = pd.read_csv('cian_flat_sale_1_2_moskovskaja oblast_28_Oct_2024_10_*.csv')
      df_rostov = pd.read_csv('cian_flat_sale_1_2_rostov-na-donu_28_Oct_2024_10_*.csv')
      df_rostov_oblast = pd.read_csv('cian_flat_sale_1_2_rostovskaja oblast_28_Oct_2024_10_*.csv')
      df_krasnodar = pd.read_csv('cian_flat_sale_1_2_krasnodar_28_Oct_2024_10_*.csv')
      df_krasnodar_oblast = pd.read_csv('cian_flat_sale_1_2_krasnodarskij kraj_28_Oct_2024_10_*.csv')

      combined_df = pd.concat([df_moscow, df_moscow_oblast, df_rostov, df_rostov_oblast, df_krasnodar, df_krasnodar_oblast], ignore_index=True)
      combined_df.to_csv('COMBINED.csv', index=False)

      print('Колво данных: ' + len(data))

get()

# import cianparser
# import pandas as pd

# def get():
#     locations = [
#         'Москва',
#         'Московская область',
#         'Ростов-на-Дону',
#         'Ростовская область',
#         'Краснодар',
#         'Краснодарский край'
#     ]
    
#     data_frames = []
    
#     for location in locations:
#         parser = cianparser.CianParser(location=location)
#         # Получаем данные
#         data = parser.get_flats(deal_type="sale", rooms=(1), with_saving_csv=True, additional_settings={"start_page": 1, "end_page": 2})
        
#         # Читаем сохраненный CSV файл
#         try:
#             df = pd.read_csv(f'cian_flat_sale_1_2_{location.lower().replace(" ", "-")}_28_Oct_2024_10_*.csv')
#             data_frames.append(df)
#         except FileNotFoundError:
#             print(f"Файл для {location} не найден.")
    
#     # Объединение всех датафреймов
#     if data_frames:
#         combined_df = pd.concat(data_frames, ignore_index=True)
#         combined_df.to_csv('COMBINED.csv', index=False)
#         print("Данные успешно объединены и сохранены в 'COMBINED.csv'.")
#     else:
#         print("Нет данных для объединения.")

# get()


# import cianparser
# import pandas as pd
# from time import sleep

# def get():
#       locations = [
#           'Москва',
#           'Московская область',
#           'Ростов-на-Дону',
#           'Ростовская область',
#           'Краснодар',
#           'Краснодарский край'
#       ]

#       data_frame = []

#       sleep(80)

#       for location in locations:
#             _parser = cianparser.CianParser(location=locations)
#             data = _parser.get_flats(deal_type="sale", rooms=(1), with_saving_csv=True, additional_settings={"start_page": 1, "end_page": 2})

#             df = pd.read_csv(f'cian_flat_sale_1_2_{location.lower().replace(" ", "-")}_28_Oct_2024_10_*.csv')
#             data_frame.append(df)
      
#       if data_frame:
#            combined_df = pd.concat(data_frame, ignore_index=True)
#            combined_df.to_csv('COMBINED.csv', index=False)
#            print('Сохранение готово')
#       else:
#            print('Не сохранилось')

# get()
```
