```python
import cianparser
import pandas as pd
from time import sleep
import random
# import json

# def load_json_data(path):
#       with open(path) as json_file:
#             data = json.load(json_file)
#       return data

# json_file_path = 'proxy.json'
# proxy_list = load_json_data(json_file_path)

# proxy = proxy_list[0]

def get():
      rooms = range(1, 7)
      locations = [
            'Москва',
            'Новая Москва',
            'Московская область'
      ]

      main_data = []
      # proxy_url = f"http://{proxy['username']}:{proxy['password']}@{proxy['host']}:{proxy['port']}"



      for main_count in range(1, 21):        
            for _location in locations:
                  for _room in rooms:
                        fir = random.randint(1,51)
                        sec = random.randint(61, 101)

                        for num_strani in range(fir,sec):
                              sleep_time_value = random.randint(10, 31)
                              sleep(sleep_time_value)

                              _parcer = cianparser.CianParser(location=_location)
                              data = _parcer.get_flats(deal_type=('sale'), rooms=(_room), with_saving_csv=False, additional_settings={'start_page':num_strani, 'end_page':num_strani}, with_extra_data=True)
                              
                              if data:
                                    main_data.extend(data)
                                    
                              sleep(sleep_time_value)

            if main_data:
                  df = pd.DataFrame(main_data)
                  df.drop_duplicates(subset='id', inplace=True)
                  df.to_csv('baza.csv', index=False, encoding='utf-8')
                              

get()
```
