```python
import cianparser
import pandas as pd
from time import sleep
import random

def get():
      rooms = range(1, 6)
      locations = [
            'Москва',
            'Новая Москва',
            'Московская область'
      ]

      main_data = []

      for main_count in range(1, 2):        
            for _location in locations:
                  for _room in rooms:
                        for num_strani in range(1,101):
                              sleep_time_value = random.randint(10, 31)
                              sleep(sleep_time_value)

                              _parcer = cianparser.CianParser(location=_location)
                              data = _parcer.get_flats(deal_type=('sale'), rooms=(_room), with_saving_csv=False, additional_settings={'start_page':num_strani, 'end_page':num_strani}, with_extra_data=False)
                              
                              if data:
                                    main_data.extend(data)
                                    
                              sleep(sleep_time_value)

            if main_data:
                  df = pd.DataFrame(main_data)
                  df.drop_duplicates(subset='id', inplace=True)
                  df.to_csv('baza.csv', index=False, encoding='utf-8')                           

get()
```
